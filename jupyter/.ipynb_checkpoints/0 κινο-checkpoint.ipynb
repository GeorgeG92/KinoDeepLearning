# 1 Join all Files together

```python
def walk_through_files(path, file_extension=('.xlsx')):                # can check for both if tuple!
    for (dirpath, dirnames, filenames) in os.walk(path):
        for filename in filenames:
            if filename.endswith(file_extension): 
                yield os.path.join(dirpath, filename).replace("\\","/")        # ensure right backslashes (Windows)!
                
def generateList(directory):
    print("Generating List of documents in parent folder...")
    documentsList = []                            
    for fname in walk_through_files(directory):
        #print(fname)
        documentsList.append(fname)
    return documentsList

# Generate Dataset from individual Files

# mylist = generateList(datapath)
# finaldf = pd.DataFrame()
# for excelpart in mylist:
#     date = excelpart.split('/')[-1][:-5].replace('kino_','')           # retrieve year_month of file 
#     if date<'2010_11':           # after this date file format changes
#         df = pd.read_excel(excelpart, header = 1).sort_values(by=['Αριθμός Κλήρωσης'])   
#     else:
#         df = pd.read_excel(excelpart, header = 2).sort_values(by=['Αριθμός Κλήρωσης'])  
#     df = df[['Αριθμός Κλήρωσης', 'Ημ/νία Κλήρωσης', 'Ώρα Κλήρωσης', '1ος ', '2ος ',       # discard extra columns
#        '3ος ', '4ος ', '5ος ', '6ος ', '7ος ', '8ος ', '9ος ', '10ος ',
#        '11ος ', '12ος ', '13ος ', '14ος ', '15ος ', '16ος ', '17ος ', '18ος ',
#        '19ος ', '20ος ']]
#     finaldf = finaldf.append(df)
# finaldf.to_csv('./olesoiklhrwseis.csv')

# Load directly

finaldf = pd.read_csv('./olesoiklhrwseis.csv')
print(finaldf.shape)  
finaldf.head()
```

# 2 Feature Engineering

## 2.1 Αριθμος Κλήρωσης 

```python
labelsCols = ['1ος ', '2ος ', '3ος ', '4ος ', '5ος ', '6ος ', '7ος ', '8ος ', '9ος ', '10ος ',
       '11ος ', '12ος ', '13ος ', '14ος ', '15ος ', '16ος ', '17ος ', '18ος ', '19ος ', '20ος ']

features = finaldf[[col for col in finaldf.columns if col not in labelsCols]]
modlist = [2, 3, 5, 7, 9, 11, 13, 17, 19, 21]

def isMod0(no, mod):
    if no%mod==0:
        return 1
    else:
        return 0

for mod in modlist:
    features['Mod_'+str(mod)+'=0'] = features.apply(lambda row: isMod0(row['Αριθμός Κλήρωσης'], mod), axis=1)
    
print(features.shape)
print(features.columns)
features.head()
```

## 2.2 Date & Time

```python
def analyseHour(time):        # quarter of day
    hour = int(time.split(':')[0])
    return hour//6
    
def analyseMinutes(time):     # part of Hour
    minutes = int(time.split(':')[1])
    return minutes//5

def analyseDate(date):        # quarter of the month
    day = pd.to_datetime(date).day
    return day//7.5

#features['Ημ/νία Κλήρωσης'] = features.apply(lambda row: sanitizeDate(row['Ημ/νία Κλήρωσης']), axis=1)

features['partOfDay'] = features.apply(lambda row: analyseHour(row['Ώρα Κλήρωσης']), axis=1)
features['partOfHour'] = features.apply(lambda row: analyseMinutes(row['Ώρα Κλήρωσης']), axis=1)
features['partOfMonth'] = features.apply(lambda row: analyseDate(row['Ημ/νία Κλήρωσης']), axis=1)
features.head()
```

# 3 One-Hot Encode Categorical Features and ditch date and no

```python
featuresStripped = features.drop(columns=['Αριθμός Κλήρωσης', 'Ημ/νία Κλήρωσης', 'Ώρα Κλήρωσης'])

catCols = ['partOfDay', 'partOfHour', 'partOfMonth']   
for catCol in catCols:                                    
    featuresStripped = pd.concat([featuresStripped, pd.get_dummies(featuresStripped[catCol], prefix=catCol)],axis=1).drop([catCol], axis=1)

featuresStripped.head()
```
